```python
flag=0
while(flag==0):
  a=int(input("Enter a number for A"))
  b=int(input("Enter a number for B"))
  op=input("Enter a operater from(+,-,*,/)")
  if (op=="+"):
    print(a+b)
  if (op=="-"):
    print(a-b)
  if (op=="*"):
    print(a*b)
  if (op=="/"):
    print(a/b)
  ch=input(print("Do you want to continue Y or N ?"))
  if (ch=="n" or ch=="N"):
    flag=1
    print("Program Exit")
  elif(ch=="y" or ch=="Y"):
    flag=0
    print("Program continue")

```

```python
def input_op():
    A = int(input("Enter Your First number: "))
    B = int(input("Enter Your Second number: "))
    F = 0
    while F == 0:
        op = input("Enter your choice (+, -, *, /): ")
        if op in ('+', '-', '*', '/'):
            F = 1
            print("Right Operator Selected")
        else:
            print("Please select the correct operator")
    return [A, B, op]

def execution_fun(input_list):
    A, B, op = input_list
    if op == '+':
        C = A + B
    elif op == '-':
        C = A - B
    elif op == '*':
        C = A * B
    elif op == '/':
        C = A / B
    else:
        print("Unsupported operator")
        return None
    print("Result:", C)
    return [C, op]

def check_execution():
    while True:
        chk = input("Do you want to continue (y or n): ")
        if chk.lower() == "n":
            print("Program exited")
            return 1
        elif chk.lower() == "y":
            print("Program continued")
            return 0
        else:
            print("Invalid input. Please enter 'y' or 'n'.")

# Main program
flag = 0
while flag == 0:
    input_list = input_op()
    result = execution_fun(input_list)
    if result is None:
        continue
    flag = check_execution()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# sin function
# cos function
# custom function
X = np.linspace(1,10,100)
Y = np.sin(X)
Z = np.cos(X)
W = 2*np.sin(X)+np.cos(X)
P = np.sin(X) - 2*np.cos(X)
```

```python
# all functions output in different figure
plt.figure(1,figsize=(5,2))#activting the figure number 1
plt.plot(X,Y,'r',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("sine function")
plt.xlabel('Range')
plt.ylabel('sin(X)')
plt.grid('on')

plt.figure(2,figsize=(5,2))#activting the figure number 1
plt.plot(X,Z,'g',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("cos function")
plt.xlabel('Range')
plt.ylabel('cos(X)')
plt.grid('on')

plt.figure(3,figsize=(5,2))#activting the figure number 1
plt.plot(X,W,'b',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("custom function")
plt.xlabel('Range')
plt.ylabel('func(X)')
plt.grid('on')

plt.figure(4,figsize=(5,2))#activting the figure number 1
plt.plot(X,P,'k',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("custom function")
plt.xlabel('Range')
plt.ylabel('func(X)')
plt.grid('on')
```

```python
# comparative analysis
plt.figure(1,figsize=(5,2))#activting the figure number 1
plt.plot(X,Y,'r',linewidth=4,label="sin(X)") # 'r','g,'m','c','k','y','b'
plt.plot(X,Z,'g',linewidth=4,label='Cos(X)')
plt.plot(X,W,'b',linewidth=4,label='func1(X)')
plt.plot(X,P,'m',linewidth=4,label='func2(X)')
plt.title("comparative analysis")
plt.xlabel('Range')
plt.ylabel('function(X)')
plt.grid('on')
```
