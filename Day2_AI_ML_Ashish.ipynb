```python
import numpy as np
np.random.seed(5)
A = np.random.randint(0, 100, size=(9, 9))
print("Elements in Matrix")
print(A)

B = []

for i in range(0, A.shape[0], 1):
    temp = A[i, :].mean()
    B.append(temp)  # save the mean on the respective position of the B list
print(B)

# step 3 and 4
C = np.zeros((A.shape[0], A.shape[1]))
ZerosCount = 0
Non_zero_Count = 0

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        if A[i, j] > B[i]:
            C[i, j] = 0
            ZerosCount = ZerosCount + 1
        else:
            C[i, j] = A[i, j]
            Non_zero_Count = Non_zero_Count + 1

if Non_zero_Count + ZerosCount == A.size:
    print("Updated matrix is verified")
else:
    print("non Validated")

print("Updated Matrix")
print(C)
print("Number of non zeros:", Non_zero_Count)
print("Number of zeros:", ZerosCount)
```

```python
import numpy as np
np.random.seed(5)
matrix = np.random.randint(0, 100, size=(4,4))
print("Elements in Matrix")
print(matrix)


# Get dimensions of the original matrix
rows, cols = matrix.shape

# Create a matrix to store max values
max_matrix = np.zeros((rows//2, cols//2), dtype=int)

# Iterate over the matrix in steps of 2
for i in range(0, rows, 2):
    for j in range(0, cols, 2):
        # Extract 2x2 submatrix
        submatrix = matrix[i:i+2, j:j+2]
        print(submatrix)

        temp = submatrix
        max_value = temp.max()
        print(f"Max value: {max_value}")
        print()  # Add a blank line between submatrices

        # Store max value in max_matrix
        max_matrix[i//2, j//2] = max_value

# Print the matrix of max values
print("Matrix of max values:")
print(max_matrix)
```
