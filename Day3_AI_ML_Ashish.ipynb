```python
#display the count plot of type 1 pokemons

plt.figure(1,figsize = (12,6))
ax = sns.countplot(data = data,x = 'Type 1')
for i in ax.containers:
  ax.bar_label(i)
plt.xticks(rotation=90)
```

```python
#display the number of pokemons in each generation show it in a visual form

plt.figure(1,figsize = (12,6))
ax = sns.countplot(data = data,x = 'Generation')
for i in ax.containers:
  ax.bar_label(i)
```

```python
#compare the Type 1 pokemons w.r.t Attack capability

plt.figure(1,figsize = (15,6))
ax = sns.barplot(data = data,x = 'Type 1',y = 'Attack')
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
```

```python
#compare the Type 2 pokemons based upon their Speed capabilities

plt.figure(1,figsize = (15,6))
ax = sns.barplot(data = data,x = 'Type 2',y = 'Speed')
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
```

```python
#compare the Type 2 pokemons generation wise based upon their Speed capabilities

plt.figure(1,figsize = (15,6))
ax = sns.barplot(data = data,x = 'Type 2',y = 'Speed',hue = 'Generation',errorbar=None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
```

```python
#compare the Type 2 pokemons generation wise based upon their Speed capabilities

plt.figure(1,figsize = (15,6))
ax = sns.barplot(data = data,x = 'Type 2',y = 'Speed',hue = 'Legendary',errorbar=None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
```

```python
#separate out the legendary pokemons and non-legendary pokemons then perform the analysis on legendary and non-legendary
#pokemons separately

#compare the attack, defense and speed w.r.t their types
#compare their total strength w.r.t generation

ind1 = data['Legendary'] == True
ind2 = data['Legendary'] == False
legend_true = data[ind1]
legend_false = data[ind2]
print(legend_true.shape)
print(legend_false.shape)
```

```python
plt.figure(1,figsize = (20,6))
plt.subplot(1,2,1)
ax = sns.barplot(data = legend_true,x = 'Type 1',y = 'Speed',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Legendary Pokemon Speed Analysis")

plt.figure(1,figsize = (20,6))
plt.subplot(1,2,2)
ax = sns.barplot(data = legend_false,x = 'Type 1',y = 'Speed',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Non-Legendary Pokemon Speed Analysis")
```

```python
plt.figure(1,figsize = (20,6))
plt.subplot(1,2,1)
ax = sns.barplot(data = legend_true,x = 'Type 1',y = 'Attack',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Legendary Pokemon Attack Analysis")

plt.figure(1,figsize = (20,6))
plt.subplot(1,2,2)
ax = sns.barplot(data = legend_false,x = 'Type 1',y = 'Attack',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Non-Legendary Pokemon Attack Analysis")
```

```python
plt.figure(1,figsize = (20,6))
plt.subplot(1,2,1)
ax = sns.barplot(data = legend_true,x = 'Type 1',y = 'Defense',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Legendary Pokemon Defense Analysis")

plt.figure(1,figsize = (20,6))
plt.subplot(1,2,2)
ax = sns.barplot(data = legend_false,x = 'Type 1',y = 'Defense',errorbar = None)
for i in ax.containers:
  ax.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Non-Legendary Pokemon Defense Analysis")
```

```python
plt.figure(1,figsize = (20,6))
plt.subplot(1,2,1)
sns.barplot(data = legend_true,x = 'Total',hue = 'Generation')
plt.figure(1,figsize = (20,6))
plt.subplot(1,2,2)
sns.barplot(data = legend_false,x = 'Total',hue = 'Generation')
```
