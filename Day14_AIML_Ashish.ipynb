```python
import random
intents = {
            "greeting":["Hi! how are you?","Hello what's up"],
            "goodbye":["Great day ahead","Nice to see you"],
            "address":["Address of the college is Near Tihar Jail Janakpuri"],
            "courses":["IITM college offers 5 graduation courses and 2 postgraduation courses"],
            "Graduation":["graduation courses are : BCA, BBA , BCOM, Btech,BJMC"],
            "Postgraduation":["Post grad courses are: MBA, MCA"]
}
```

```python
def detect_response(query):
  prediction = humingbird.Text.predict(
    text = query, # inferencing text
    labels=["greeting","goodbye",
            "address","courses",
            "Graduation","Postgraduation"] # suply potential labels
  )
  max_score=0
  for val in prediction:
    if(val['score']>max_score):
      max_score=val['score']
      max_val = val['score']
      max_val_text = val['className']
  print(max_val_text)
  return random.choice(intents[max_val_text])
```

```python
import random

intents = {
    "food_calories": {
        "apple": ["An apple has about 95 calories.", "A medium apple contains approximately 95 calories."],
        "banana": ["A banana has about 105 calories.", "A medium banana contains approximately 105 calories."],
        "orange": ["An orange has about 62 calories.", "A medium orange contains approximately 62 calories."],
        "grapes": ["A cup of grapes has about 62 calories.", "One cup of grapes contains approximately 62 calories."],
        "strawberry": ["A cup of strawberries has about 49 calories.", "One cup of strawberries contains approximately 49 calories."],
        "broccoli": ["A cup of broccoli has about 55 calories.", "One cup of broccoli contains approximately 55 calories."],
        "chicken_breast": ["A grilled chicken breast has about 165 calories.", "A medium grilled chicken breast contains approximately 165 calories."],
        "salmon": ["A 3-ounce serving of salmon has about 180 calories.", "A 3-ounce serving of salmon contains approximately 180 calories."],
        "egg": ["A large egg has about 78 calories.", "One large egg contains approximately 78 calories."],
        "rice": ["A cup of cooked rice has about 206 calories.", "One cup of cooked rice contains approximately 206 calories."],
        "almond": ["A quarter-cup of almonds has about 200 calories.", "A quarter-cup of almonds contains approximately 200 calories."],
        "burger": ["A regular beef burger has about 250-300 calories.", "A standard beef burger contains approximately 250-300 calories."],
        "pizza": ["A slice of pizza has about 285 calories.", "One slice of pizza contains approximately 285 calories."],
        "french_fries": ["A medium serving of french fries has about 365 calories.", "A medium portion of french fries contains approximately 365 calories."],
        "soda": ["A can of soda has about 140 calories.", "One can of soda contains approximately 140 calories."]
    }
}

def get_food_calories(food_name):
    if food_name in intents["food_calories"]:
        return random.choice(intents["food_calories"][food_name])
    else:
        return "I don't have calorie information for that food."

# Example usage
food_name = input("Enter Your Food To Know About Calaories")
print(get_food_calories(food_name))
```
